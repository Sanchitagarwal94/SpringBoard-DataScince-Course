#### Applying data wrangling techniques on train dataset

```python
# Creating a copy of train file to apply data wrangling operations and changing the index
train_copy = train.copy()
```

```python
#dropping the duplicates from the file if any:
train_copy=train_copy.drop_duplicates().dropna()
```
